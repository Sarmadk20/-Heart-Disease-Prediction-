```python
# One hot encoding categorical columns
df = pd.get_dummies(df)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_recall_curve,roc_curve,auc,RocCurveDisplay, PrecisionRecallDisplay
```

```python
X = df.drop('target',axis=1)
y = df['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(scaled_X_train,y_train)
```

```python
pred = model.predict(scaled_X_test)
print(classification_report(y_test,pred))
```

```python
from sklearn.linear_model import LogisticRegressionCV 
model = LogisticRegressionCV()

model.fit(scaled_X_train,y_train)
```

```python
preds = model.predict(scaled_X_test)
```

```python

print(classification_report(y_test,preds))
```

```python
# Imports the Pipeline class from scikit-learn.
# Defines a pipeline with a single operation: logistic regression

from sklearn.pipeline import Pipeline
operations  = [('log_model',LogisticRegression())]
pipe = Pipeline(operations)
```

```python
# Setting up possible values for logistic regression hyperparameters: solver types, penalties, l1_ratio (for elasticnet), and regularization strength (C) on a log scale.
solver = ['saga']
penalty = ['l1','l2','elasticnet']
l1_ratio = np.linspace(0,5,10)
C = np.logspace(0, 10, 20)
```

```python
# Building the parameter grid for hyperparameter search, mapping parameter names to possible values. The double underscore __ syntax references parameters inside the pipeline.
param_grid = {
    'log_model__solver': solver,
    'log_model__penalty': penalty,
    'log_model__l1_ratio': l1_ratio,  
    'log_model__C': C
}
```

```python
# Initializing grid search cross-validation over the pipeline using defined parameters, evaluating with 5-fold CV based on accuracy.
grid_model = GridSearchCV(pipe,param_grid,cv=5,scoring='accuracy')
```

```python
# Trains the model using grid search with the scaled training data and target.
grid_model.fit(scaled_X_train,y_train)
```

```python
# Retrieves the set of hyperparameters that produced the best accuracy. 
grid_model.best_params_
```

```python
# Uses the trained, best-selected model to make predictions on the scaled test set.
pred = grid_model.predict(scaled_X_test)
```

```python

print(classification_report(y_test,pred))
```

```python
cm = confusion_matrix(y_test,pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```

```python
fpr,tpr,thresholds = roc_curve(y_test,pred)

roc_auc= auc(fpr,tpr)
print("AUC:", roc_auc)
```

```python
RocCurveDisplay.from_estimator(grid_model, scaled_X_test, y_test)
```

## Final Insights
- Logistic Regression provides a simple, interpretable baseline model.  
- The dataset effectively predicts heart disease risk using measurable medical attributes.  
- The model can be improved using:
  - Additional models (Random Forest, XGBoost)
  - Handling missing values more robustly
    
- Good Model Performance:
The AUC score shown on the graph is 0.87, and printed AUC is ~0.79, both indicating that the model has a good ability to distinguish between the positive and negative classes. Typically, an AUC above 0.8 is considered strong model performance.

- Low False Positive Rate for High True Positive Rate:

The ROC curve rises quickly towards the top-left, which means the model achieves a high true positive rate (correctly identifying positives) while keeping the false positive rate (incorrectly labeling negatives as positives) relatively low. This is a desirable outcome in many classification tasks.

Overall, the notebook demonstrates a complete pipeline from data loading to evaluation with clear predictive insights.
